```python
from PIL import Image
import matplotlib.pyplot as plt

# Cargamos las imgenes para visualizar y determinar la altura de la marca de agua
images_paths = [
    "imgs/1.jpg",
    "imgs/2.jpg",
    "imgs/3.jpg"
]

# Ajustaremos el cdigo para mostrar la imagen original y la imagen recortada una al lado de la otra para comparacin

# Vamos a mostrar las imgenes y sus ltimas filas para estimar el tamao de la marca de agua
for image_path in images_paths:
    img = Image.open(image_path)
    
    # Obtenemos las dimensiones de la imagen original
    width, height = img.size
    
    # Definimos los lmites del recorte para eliminar los ltimos 50 pxeles
    cropped_img = img.crop((0, 0, width, height - 50))  # Recortamos los ltimos 50 pxeles
    
    # Mostramos la imagen original y la recortada una al lado de la otra
    plt.figure(figsize=(10, 5))  # Tamao aumentado para mejor visualizacin

    # Imagen original
    plt.subplot(1, 2, 1)
    plt.imshow(img)
    plt.title(f'Original - {image_path}')
    plt.axis('off')  # Ocultamos los ejes para mejor claridad
    
    # Imagen recortada
    plt.subplot(1, 2, 2)
    plt.imshow(cropped_img)
    plt.title(f'Recortada - {image_path}')
    plt.axis('off')  # Ocultamos los ejes para mejor claridad

# Mostramos las imgenes
plt.show()
```
